```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import zeyrek
import nltk
from nltk.corpus import stopwords

# 1. HazÄ±rlÄ±k: Zeyrek ve Stopwords
analyzer = zeyrek.MorphAnalyzer()
nltk.download('stopwords')
stop_words_tr = set(stopwords.words('turkish'))
# Metne Ã¶zel temizlenmesi gereken ek kelimeler
ekstra_stop = {"bir", "bizi", "mi", "deÄŸil", "ise", "peki", "yani", "kendi", "mÄ±", "ancak", "ÅŸÃ¶yle", "bÃ¶yle"}
stop_words_tr.update(ekstra_stop)

# 2. Metin (Senin paylaÅŸtÄ±ÄŸÄ±n uzun metin)
metin = """
Yapay zekÃ¢ artÄ±k sadece sorularÄ±mÄ±za cevap vermiyor.
Bizi dinliyor, Ã¶nerilerde bulunuyor, duygularÄ±mÄ±za uygun cÃ¼mleler kuruyor...
(Metnin tamamÄ±nÄ± buraya tÄ±rnaklar iÃ§ine yapÄ±ÅŸtÄ±rabilirsin)
"""

# 3. KÃ¶k Bulma (Zeyrek) ve Temizleme Fonksiyonu
def metni_temizle_ve_kok_bul(text):
    kelimeler = text.lower().split()
    temiz_metin = []
    for kelime in kelimeler:
        # Zeyrek ile kÃ¶k bul (Ä°lk sonucu alÄ±yoruz)
        analiz = analyzer.lemmatize(kelime)
        kok = analiz[0][1][0] if analiz else kelime
        # Stopwords deÄŸilse ve harf ise ekle
        if kok not in stop_words_tr and len(kok) > 2:
            temiz_metin.append(kok)
    return " ".join(temiz_metin)

# 4. Ä°ÅŸlemi BaÅŸlat
temiz_icerik = metni_temizle_ve_kok_bul(metin)

# 5. TF-IDF Analizi
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform([temiz_icerik])

# 6. SonuÃ§larÄ± Tabloya DÃ¶k
kelimeler = vectorizer.get_feature_names_out()
skorlar = tfidf_matrix.toarray()[0]

anahtar_kelimeler = pd.DataFrame({'Kelime': kelimeler, 'Ã–nem Skoru': skorlar})
print("\nðŸš€ Metnin En Ã–nemli Anahtar Kelimeleri:")
print(anahtar_kelimeler.sort_values(by='Ã–nem Skoru', ascending=False).head(10))
```
