```python
import pandas as pd
import random
import warnings
from zeyrek import MorphAnalyzer
import nltk

# HazÄ±rlÄ±k
warnings.filterwarnings("ignore")
analyzer = MorphAnalyzer()
nltk.download('stopwords', quiet=True)
durak_kelimeler = nltk.corpus.stopwords.words('turkish')

# 1. GERÃ‡EKÃ‡Ä° VERÄ° SETÄ° (3000 SatÄ±r)
senaryolar = [
    ("Kargo Ã§ok yavaÅŸ geldi, paket ezilmiÅŸti.", 1),
    ("ÃœrÃ¼n kalitesi muazzam, fiyatÄ±nÄ± hak ediyor.", 5),
    ("MÃ¼ÅŸteri hizmetleri ilgisiz, iade sÃ¼reci Ã§ok zor.", 2),
    ("HÄ±zlÄ± teslimat iÃ§in teÅŸekkÃ¼rler, Ã§ok ÅŸÄ±k bir Ã¼rÃ¼n.", 5),
    ("Fiyat performans Ã¼rÃ¼nÃ¼, tavsiye ederim.", 4),
    ("KumaÅŸÄ± Ã§ok ince, beklentimi karÅŸÄ±lamadÄ±.", 2),
    ("ParamparÃ§a bir paket, kurye Ã§ok kaba.", 1),
    ("Harika bir alÄ±ÅŸveriÅŸti, hÄ±zlÄ± ve gÃ¼venli.", 5)
]

data = {
    'tarih': pd.date_range(start='2024-01-01', periods=3000, freq='H'),
    'yorum': [random.choice(senaryolar)[0] for _ in range(3000)],
    'puan': [random.choice(senaryolar)[1] for _ in range(3000)]
}
df_proje = pd.DataFrame(data)

# 2. PROFESYONEL TEMÄ°ZLÄ°K PÄ°PELÄ°NE'I
def nlp_pipeline(metin):
    # Temizlik
    metin = "".join([c.lower() for c in metin if c.isalpha() or c.isspace()])
    # KÃ¶k Bulma & Stopwords
    temiz = []
    for kelime in metin.split():
        if kelime not in durak_kelimeler:
            analiz = analyzer.lemmatize(kelime)
            temiz.append(analiz[0][1][0])
    return " ".join(temiz)

print("Veriler iÅŸleniyor (NLP Pipeline)...")
df_proje['temiz_yorum'] = df_proje['yorum'].apply(nlp_pipeline)

# 3. Ä°ÅžE ALDIRACAK ANALÄ°Z: KRÄ°TÄ°K ÅžÄ°KAYET TESPÄ°TÄ°
print("\n--- ANALÄ°Z RAPORU ---")

# DÃ¼ÅŸÃ¼k puanlÄ± yorumlarda en Ã§ok hangi kelimeler geÃ§iyor?
negatif_kelimeler = " ".join(df_proje[df_proje['puan'] <= 2]['temiz_yorum']).split()
sikayet_odaklari = pd.Series(negatif_kelimeler).value_counts().head(5)

print("\nðŸš¨ ACÄ°L Ã‡Ã–ZÃœLMESÄ° GEREKEN SORUNLAR (Olumsuz Odaklar):")
print(sikayet_odaklari)

# Ortalama MÃ¼ÅŸteri Memnuniyeti
print(f"\nâ­ Genel Memnuniyet Skoru: {df_proje['puan'].mean():.2f} / 5.00")
```
